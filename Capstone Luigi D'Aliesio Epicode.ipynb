```python
import pandas as pd
```

```python
df = pd.read_csv('Titanic-Original.csv', sep=';',
                 index_col='PassengerId',
                 dtype={'Name': pd.StringDtype(),
                'Sex' : pd.StringDtype(), 
                'Ricket' : pd.StringDtype(), 
            'Cabin' : pd.StringDtype(), 
             'Embarked' : pd.StringDtype()})
```

```python
df.info()
```

```python
# descrizione statistica delle colonne numeriche 
df.describe()
```

```python
# vediamo che ha acquistato i biglietti più costosi
df.sort_values('Fare')
```

```python
df.sort_values('Fare', ascending=False)
```

```python
# eliminiamo valori nulli nella colonna età
df.dropna(subset='Age')
```

```python
# grafico età passeggeri
df['Age'].plot()
```

```python
# grafico per vedere su che classe viaggiavano maggiormente i passeggeri
df['Pclass'].value_counts().plot(kind='pie')
```

```python
# grafico sopravvissuti 
df['Survived'].value_counts().plot(kind='pie')
```

```python
# reaggruppare tutti i passeggeri della prima classe
df.groupby('Pclass').get_group(1)
```

```python
# tasso di sopravvivenza medio per classe
df.groupby('Pclass')['Survived'].mean().plot(kind='bar')
```
